```python
#Data understanding
import pandas as pd
```

```python
df = pd.read_csv('e commerce dataset.csv',encoding='latin1')
df.head()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df.info()
```

```python
df.shape
```

```python
df.describe()
```

```python
#Analisis Monthly Sales Tren
df['Order month']= df['Order Date'].dt.month  # making new order month  columns 
df['Order year']= df['Order Date'].dt.year  # making new order year columns
```

```python
df.head()
```

```python
monthly_sales = df.groupby('Order month')['Sales'].sum().reset_index()
monthly_sales.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(monthly_sales, x='Order month', y='Sales', marker='o')
plt.xlabel('Order Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Analysis')
plt.show()
```

```python
monthly_sales.describe()
```

```python
#Top Product by Revenue and Quantity
#Revenue
top_revenue_products = df.groupby('Product Name')['Sales'].sum().reset_index()
top_revenue_products = top_revenue_products.sort_values(by='Sales', ascending=False).head(10)
top_revenue_products
```
